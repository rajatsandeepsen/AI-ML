```python
list = [1, 2, 3, 4, 5]
print(list)
```

iterating using for loop:


```python
for i in range(len(list)):
    print(list[i], end=" ")
print("\n")
```

The values of the list are :


```python
while i < len(list):
    print(list[i], end=" ")
    i += 1
print("\n")
```

```python
print("Iterating using comprehension :", [i for i in list])
```

Iterating using enumerate:


```python
for i, item in enumerate(list):
    print(i, item)
print("\n")
```

Adding an element to list:


```python
list.append(6)
print(list)
```

Remove an element from a list by index


```python
list.pop(0)
print(list)
```

Remove an element from a list by value


```python
list.remove(3)
print(list)
```

```python
list.reverse()
print("Reversed list is :", list)
```

```python
list.sort()
print("Sorted list is :", list)
```

```python
count = list.count(2)
print("The number of occurance of 2 is :", count)
```

```python
index = list.index(4)
print("Index of element 4 is :", index)
```

```python
min_element = min(list)
print("Minimum element is :", min_element)
```

```python
max_element = max(list)
print("Maximum element is :", max_element)
```

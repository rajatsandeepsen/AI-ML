```python
# Volume & Surface Area of a Cylinder

from math import pi

r = int(input("Enter Radius "))
h = int(input("Enter Height "))

vol = pi * r * r * h
sf = 2 * pi * r * (r + h)

print("Volume", vol)
print("Surface Area", sf)
```

```python
import re

def check(string):
    regex = "[a-zA-Z0-9-_]+@[a-zA-Z0-9]+\.[a-z]{1,3}$"
    if re.match(regex, string):
        return True
    
    return False

email = input("Enter Email ")
print(check(email))
```

Adding an element to list:


```python
# Matrix Multiplication

row = int(input("Enter row "))
col = int(input("Enter col "))

matrix = []
matrix2 = []

for r in range(0,row):
    w = []
    for c in range(0,col):
        string = "[" + str(r) + "][" + str(c) + "] = "
        w.append(int(input(string)))
    matrix.append(w)

    
for r in range(0,row):
    w = []
    for c in range(0,col):
        string = "[" + str(r) + "][" + str(c) + "] = "
        w.append(int(input(string)))
    matrix2.append(w)

sum = []
for r in range(0,row):
    w = []
    for c in range(0,col):
        w.append(matrix[r][c] + matrix2[r][c])
    sum.append(w)

for i in sum:
    print(i)
```

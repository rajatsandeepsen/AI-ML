```python
dist = {"A":"Apple","B":"Ball","C":"Cat"}

print(dist.keys())
print(dist.values())

dist['D'] = 'Dog'
print(dist)
```

```python
for key, value in dist.items():
    dist[key] = value + value
    
print(dist)
```

```python
x = int(input("Enter Tuple Size "))
tuple = ()
for i in range(x):
    t = int(input("Enter value "))
    tuple += (t,)

even = ()
odd = ()

for i in tuple:
    if (i % 2) == 0:
       even += (i,) 
    else:
        odd += (i,)

print("\nOrginal Tuple", tuple)
print("Even only", even)
print("Odd only", odd)
```

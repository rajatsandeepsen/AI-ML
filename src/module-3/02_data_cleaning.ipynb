# 02 - Data Cleaning

Goal: Find missing values and apply simple cleaning steps.


```python
import pandas as pd
```

```python
df = pd.read_csv('src/module-3/datasets/student_lifestyle.csv')
df.isnull().sum()
```

```python
# Fill numeric missing values with median
numeric_cols = ['study_hours', 'sleep_hours', 'daily_screen_time', 'attendance_percent', 'exam_score']
for col in numeric_cols:
	df[col] = df[col].fillna(df[col].median())

df.isnull().sum()
```

```python
# Basic transformation: create score category
df['score_level'] = pd.cut(df['exam_score'], bins=[0, 60, 75, 90, 100], labels=['Low', 'Average', 'Good', 'Excellent'])
df[['student_id', 'exam_score', 'score_level']].head()
```

```python
# Save cleaned file
df.to_csv('src/module-3/datasets/student_lifestyle_cleaned.csv', index=False)
```

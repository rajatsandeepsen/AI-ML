# 03 - Visualization Basics

Goal: Create simple charts with Matplotlib and Seaborn.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('src/module-3/datasets/student_lifestyle_cleaned.csv')
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['exam_score'], bins=8, kde=True)
plt.title('Distribution of Exam Score')
plt.xlabel('Exam Score')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='study_hours', y='exam_score', hue='gender')
plt.title('Study Hours vs Exam Score')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='gender', y='sleep_hours')
plt.title('Sleep Hours by Gender')
plt.show()
```
